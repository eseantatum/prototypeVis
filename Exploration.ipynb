```python
import json
import pandas as pd
#from shapely.geometry import mapping


with open("Raw File/alleghenycounty_streetcenterlines202202.geojson") as f:
    streets = json.load(f)
    
print(streets.keys())
print(streets['type'][0])
print(streets['features'][0])
print(streets['features'][0].keys())

s = streets['features']
```

```python
print(len(s))

df = pd.DataFrame.from_dict(s)
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'index'})
display(df)
```

```python
#display(df['properties'])
prop = df['properties']

p = pd.DataFrame.from_dict(prop.tolist())
p.reset_index(inplace=True)
p = p.rename(columns = {'index':'index'})
#display(p)

#df.merge(p, how="right", on="index")
#display(df)

df = pd.concat([df,p], axis=1, join='inner')
#df.drop("properties")
#display(df)
#df = df.drop('ST_NAME')
display(df)
```

```python
#display(df['properties'])
geo = df['geometry']

g = pd.DataFrame.from_dict(geo.tolist())
g.reset_index(inplace=True)
g = g.rename(columns = {'index':'index'})
#display(g)

#df.merge(g, how="right", on="index")
#display(df)

df = pd.concat([df,g], axis=1, join='inner')
#df.drop("properties")
#display(df)
#display(df.columns)

dfsimplified = df[["id", 'index', 'geometry', 'properties', 'ST_NAME','ST_TYPE','ST_NAME', 'LCOUNTY','SPEED', 'ST_PREFIX','ST_SUFFIX','coordinates' ]]
display(dfsimplified)
```

```python
coordinates = dfsimplified['coordinates']



for i in range(10):
    display(coordinates[i])
```

```python
dfsimplified['color'] = 0

display(dfsimplified)

```

```python
point = (-79.7789, 40.44444)

i = 0
row = 0
distance = (100,100)
flag = False

for coordinateList in coordinates:
    for p in coordinateList:
        currentDis = (abs(point[0] - p[0]), abs(point[1] - p[1]))
        #print(currentDis)
        if((currentDis[0] + currentDis[1]) <= (distance[0] + distance[1])):
            distance = currentDis
            flag = True
        #print(distance)
        #print("----------------------------------------")
    if(flag):
        flag = False
        row = i
    i+=1

#print(row)
dfsimplified['color'][row] = 1
#for x in range(100):
#    dfsimplified['color'][x] = 1
display(dfsimplified['color'][row])
#display(dfsimplified)
#dfsimplified=dfsimplified.drop(columns='index')
dfsimplified = dfsimplified.loc[:,~dfsimplified.columns.duplicated()]
display(dfsimplified)

        
```
